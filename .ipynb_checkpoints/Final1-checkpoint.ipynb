```python
import numpy as np
import math
import PIL
from matplotlib.pyplot import imshow
```

```python
def get_meta_key(img_arr):
    
    img_arr = img_arr / 255
    meta_key = 0
    
    for i in range(img_arr.shape[0]):
        for j in range(img_arr.shape[1]):
            for k in range(img_arr.shape[2]):
                meta_key += img_arr[i,j,k]
                
    return round((meta_key / (img_arr.shape[0] * img_arr.shape[1] * img_arr.shape[2])), 6)
```

```python
def keygen(x, r, size):
    
    key = []
    for i in range(size):
        x = x*r*(1-x)
        key.append(int((x * math.pow(10,16))% 256))
        
    key = np.array(key)
    key = np.reshape(key, (img_arr.shape[0], img_arr.shape[1], img_arr.shape[2]))
    
    return key
```

```python
def Lorenzkey(x0, y0, z0, num_steps):
    
    dt = 0.01
    xs = np.empty(num_steps + 1)
    ys = np.empty(num_steps + 1)
    zs = np.empty(num_steps + 1)
    xs[0], ys[0], zs[0] = x0, y0, z0
    
    s = 10
    r = 28
    b = 8/3
    
    for i in range(num_steps):
        xs[i+1] = (xs[i] + (s * ys[i] -xs[i] * dt)) % 256
        ys[i+1] = (ys[i] + ((xs[i] * (r - zs[i]) - ys[i]) * dt)) % 256
        zs[i+1] = (zs[i] + ((xs[i] * ys[i] - b * zs[i]) * dt)) % 256
        
    return xs, ys, zs
```

```python
def decimalToBinary(n):
    
    bin_n = bin(n).replace("0b", "")
    
    if len(bin_n) < 8:
        temp_0 = ""
        for i in range(8 - len(bin_n)):
            temp_0 += '0'
            
        bin_n = temp_0 + bin_n
    
    return bin_n
```

```python
def decToBinary_dims(key):

    first_dim = []
    second_dim = []
    third_dim = []

    for i in range(key.shape[0]):
        temp_li1 = []
        temp_li2 = []
        temp_li3 = []

        for j in range(key.shape[1]):
            
            temp_li1.append(decimalToBinary(key[i,j,0]))
            temp_li2.append(decimalToBinary(key[i,j,1]))
            temp_li3.append(decimalToBinary(key[i,j,2]))

        first_dim.append(temp_li1)
        second_dim.append(temp_li2)
        third_dim.append(temp_li3)
    
    return first_dim, second_dim, third_dim
```

```python
def get_mat(img_arr, encoding_rule1, first_dim, second_dim, third_dim):
    mat = np.empty(img_arr.shape)

    for i in range(img_arr.shape[0]):
        for j in range(img_arr.shape[1]):

            last2_current = first_dim[i][j][6:]
            first2_next = first_dim[i][(j+1)%img_arr.shape[1]][:2]

            if (encoding_rule1[last2_current] == 'A' and encoding_rule1[first2_next] == 'T'):
                mat[i][j][0] = 1
            elif (encoding_rule1[last2_current] == 'T' and encoding_rule1[first2_next] == 'A'):
                mat[i][j][0] = 1
            elif (encoding_rule1[last2_current] == 'C' and encoding_rule1[first2_next] == 'G'):
                mat[i][j][0] = 1
            elif (encoding_rule1[last2_current] == 'G' and encoding_rule1[first2_next] == 'C'):
                mat[i][j][0] = 1
            else:
                mat[i][j][0] = 0

    for i in range(img_arr.shape[0]):
        for j in range(img_arr.shape[1]):

            last2_current = second_dim[i][j][6:]
            first2_next = second_dim[i][(j+1)%img_arr.shape[1] ][:2]

            if (encoding_rule1[last2_current] == 'A' and encoding_rule1[first2_next] == 'T'):
                mat[i][j][1] = 1
            elif (encoding_rule1[last2_current] == 'T' and encoding_rule1[first2_next] == 'A'):
                mat[i][j][1] = 1
            elif (encoding_rule1[last2_current] == 'C' and encoding_rule1[first2_next] == 'G'):
                mat[i][j][1] = 1
            elif (encoding_rule1[last2_current] == 'G' and encoding_rule1[first2_next] == 'C'):
                mat[i][j][1] = 1
            else:
                mat[i][j][1] = 0

    for i in range(img_arr.shape[0]):
        for j in range(img_arr.shape[1]):

            last2_current = third_dim[i][j][6:]
            first2_next = third_dim[i][(j+1)%img_arr.shape[1] ][:2]

            if (encoding_rule1[last2_current] == 'A' and encoding_rule1[first2_next] == 'T'):
                mat[i][j][2] = 1
            elif (encoding_rule1[last2_current] == 'T' and encoding_rule1[first2_next] == 'A'):
                mat[i][j][2] = 1
            elif (encoding_rule1[last2_current] == 'C' and encoding_rule1[first2_next] == 'G'):
                mat[i][j][2] = 1
            elif (encoding_rule1[last2_current] == 'G' and encoding_rule1[first2_next] == 'C'):
                mat[i][j][2] = 1
            else:
                mat[i][j][2] = 0
                
    return mat
```

```python
def get_dims(img_arr, mat):
    
    dim1_0 = []
    dim2_0 = []
    dim3_0 = []
    
    for i in range(img_arr.shape[0]):
        
        temp1_0 = 0
        temp2_0 = 0
        temp3_0 = 0
        
        for j in range(img_arr.shape[1]):
            if mat[i,j,0] == 0:
                temp1_0 += 1
                
            if mat[i,j,1] == 0:
                temp2_0 += 1

            if mat[i,j,2] == 0:
                temp3_0 += 1

        dim1_0.append(temp1_0)
        dim2_0.append(temp2_0)
        dim3_0.append(temp3_0)
        
        
    dim1_0 = np.array(dim1_0) / 255
    dim2_0 = np.array(dim2_0) / 255
    dim3_0 = np.array(dim3_0) / 255

    return dim1_0, dim2_0, dim3_0
```

```python
def make_gcd_matrix(key):
    
    pix_val_key = []
    
    for i in range(key.shape[0]):
        for j in range(key.shape[1]):
            pix_val_key.append((key[i,j,0], key[i,j,1], key[i,j,2]))
    
    rows = key.shape[0]
    cols = key.shape[1]
    
    gcd_matrix = []

    temp_li = []
    for pix_val in pix_val_key:
        gcd1 = math.gcd(pix_val[0], pix_val[1])
        gcd2 = math.gcd(pix_val[1], pix_val[2])
        gcd3 = math.gcd(pix_val[2], pix_val[0])
        gcd = (gcd1, gcd2, gcd3)
        temp_li.append(gcd)
        if len(temp_li) % key.shape[0] == 0:
            if len(temp_li) == 0:
                temp_li = []
            else:
                gcd_matrix.append(temp_li)
                temp_li = []
          
    return gcd_matrix

def get_sum_mod(gcd_matrix):
    
    sum_mod = []
    for i in range(len(gcd_matrix)):
        temp_sum_mod1 = 0
        temp_sum_mod2 = 0
        temp_sum_mod3 = 0
        for j in range(len(gcd_matrix[i])):
            if i == j:
                continue
            temp_sum_mod1 += gcd_matrix[i][j][0]
            temp_sum_mod2 += gcd_matrix[i][j][1]
            temp_sum_mod3 += gcd_matrix[i][j][2]
            
        temp_sum_mod = [temp_sum_mod1 % 256, temp_sum_mod2 % 256, temp_sum_mod3 % 256]
        
        sum_mod.append(temp_sum_mod)
    return sum_mod

def trans(sum_mod):
    
    trans_mod = []
    for i in range(len(sum_mod)):
        temp = []
        for j in range(len(sum_mod[i])):
            temp.append(sum_mod[i][j] % len(sum_mod))
        trans_mod.append(temp)
        
    dim1 = []
    dim2 = []
    dim3 = []
    
    for i in range(len(trans_mod)):
        if trans_mod[i][0] not in dim1:
            dim1.append(trans_mod[i][0])
        
        if trans_mod[i][0] in dim1:
            for j in range(trans_mod[i][0]+1, len(trans_mod)):
                if j not in dim1:
                    dim1.append(j)
                    break
                if j == len(trans_mod) -1:
                    for k in range(0, trans_mod[i][0]):
                        if k not in dim1:
                            dim1.append(k)
                            break
                            
        if trans_mod[i][1] not in dim2:
            dim2.append(trans_mod[i][1])
            
        if trans_mod[i][1] in dim1:
            for j in range(trans_mod[i][1]+1, len(trans_mod)):
                if j not in dim2:
                    dim2.append(j)
                    break
                if j == len(trans_mod) -1:
                    for k in range(0, trans_mod[i][1]):
                        if k not in dim2:
                            dim2.append(k)
                            break

        if trans_mod[i][2] not in dim3:
            dim3.append(trans_mod[i][2])
            
        if trans_mod[i][2] in dim3:
            for j in range(trans_mod[i][2]+1, len(trans_mod)):
                if j not in dim3:
                    dim3.append(j)
                    break
                if j == len(trans_mod) -1:
                    for k in range(0, trans_mod[i][2]):
                        if k not in dim3:
                            dim3.append(k)
                            break
                            
    return dim1, dim2, dim3

def encrypt_trans(dim1, dim2, dim3, subs_cipher_img):
    
    cipher_img = np.empty(subs_cipher_img.shape, dtype = 'uint8')
    
    for i in range(subs_cipher_img.shape[0]):
        for a,b,c in zip(range(len(dim1)), range(len(dim2)), range(len(dim3))):
            index1 = dim1[a]
            index2 = dim2[b]
            index3 = dim3[c]
            
#             k = 0
#             while index1 + k * len(dim1) < cipher_img.shape[1]:
            cipher_img[i,a , 0] = subs_cipher_img[i, index1 , 0]
#                 k += 1
                
#             k = 0    
#             while index2 + k * len(dim2) < cipher_img.shape[1]:
            cipher_img[i,b , 1] = subs_cipher_img[i, index2 , 1]
#                 k += 1
                
#             k = 0
#             while index3 + k * len(dim3) < cipher_img.shape[1]:
            cipher_img[i,c , 2] = subs_cipher_img[i, index3, 2]
#                 k += 1
            
    return cipher_img


def encrypt_trans1(dim1, dim2, dim3, subs_cipher_img):
    
    cipher_img = np.empty(subs_cipher_img.shape, dtype = 'uint8')
    
    for j in range(subs_cipher_img.shape[1]):
        
        for a,b,c in zip(range(len(dim1)), range(len(dim2)), range(len(dim3))):
            index1 = dim1[a] 
            index2 = dim2[b]
            index3 = dim3[c]
            
#             k = 0
#             while index1 + k * len(dim1) < cipher_img.shape[0]:
            cipher_img[a,j, 0] = subs_cipher_img[index1, j, 0]
#                 k += 1
                
#             k = 0
#             while index2 + k * len(dim2) < cipher_img.shape[0]:
#                 print(index2, k)
            cipher_img[b,j, 1] = subs_cipher_img[index2, j, 1]
#                 k += 1
                
#             k = 0
#             while index3 + k * len(dim3) < cipher_img.shape[0]:
#                 print(c + k * len(dim3))
            cipher_img[c,j, 2] = subs_cipher_img[index3, j, 2]
#                 k += 1
            
    return cipher_img

def decrypt_trans1(dim1, dim2, dim3, cipher_img):
        
    plain_img = np.empty(cipher_img.shape, dtype = 'uint8')
    
    for j in range(cipher_img.shape[1]):
        for a,b,c in zip(range(len(dim1)), range(len(dim2)), range(len(dim3))):
            index1 = dim1[a]
            index2 = dim2[b]
            index3 = dim3[c]
            
#             k = 0
#             while index1 + k * len(dim1) < cipher_img.shape[0]:
            plain_img[index1, j, 0] = cipher_img[a,j,0]
#                 k += 1
                
#             k = 0
#             while index2 + k * len(dim2) < cipher_img.shape[0]:
            plain_img[index2, j, 1] = cipher_img[b,j,1]
#                 k += 1
                
#             k = 0
#             while index3 + k * len(dim3) < cipher_img.shape[0]:
            plain_img[index3, j, 2] = cipher_img[c,j,2]
#                 k += 1
            
    return plain_img

def decrypt_trans(dim1, dim2, dim3, cipher_img):
        
    plain_img = np.empty(cipher_img.shape, dtype = 'uint8')
    
    for i in range(cipher_img.shape[0]):
        for a,b,c in zip(range(len(dim1)), range(len(dim2)), range(len(dim3))):
            index1 = dim1[a]
            index2 = dim2[b]
            index3 = dim3[c]
            
#             k = 0
#             while index1 + k * len(dim1) < cipher_img.shape[1]:
            plain_img[i, index1, 0] = cipher_img[i,a, 0]
#                 k += 1
            
#             k = 0
#             while index2 + k * len(dim2) < cipher_img.shape[1]:
            plain_img[i, index2, 1] = cipher_img[i,b, 1]
#                 k += 1
                
#             k = 0
#             while index3 + k * len(dim3) < cipher_img.shape[1]:
            plain_img[i, index3, 2] = cipher_img[i,c, 2]
#                 k += 1
            
    return plain_img
```

```python
def encrypt_image(img_arr, r, x):
    
    encoding_rule1 = {'00' : 'A', '11' : 'T', '01' : 'C', '10' : 'G'}
    encoding_rule2 = {'00' : 'A', '11' : 'T', '01' : 'G', '10' : 'C'}
    encoding_rule3 = {'00' : 'C', '11' : 'G', '01' : 'A', '10' : 'T'}
    encoding_rule4 = {'00' : 'G', '11' : 'C', '01' : 'A', '10' : 'T'}
    encoding_rule5 = {'10' : 'A', '01' : 'T', '00' : 'C', '11' : 'G'}
    encoding_rule6 = {'10' : 'A', '01' : 'T', '11' : 'C', '00' : 'G'}
    encoding_rule7 = {'11' : 'A', '00' : 'T', '01' : 'C', '10' : 'G'}
    encoding_rule8 = {'11' : 'A', '00' : 'T', '10' : 'C', '01' : 'G'}
    
    meta_key = get_meta_key(img_arr)
    print(meta_key)
    key = keygen(meta_key, r, img_arr.shape[0] * img_arr.shape[1] * img_arr.shape[2])
#     key = np.reshape(key, (img_arr.shape[0], img_arr.shape[1], img_arr.shape[2]))
    gcd_matrix = make_gcd_matrix(key)
    sum_mod = get_sum_mod(gcd_matrix)

    dim1, dim2, dim3 = trans(sum_mod)
    cipher_img_trans = img_arr.copy()
    
#     for _ in range(3):
    cipher_img_trans = encrypt_trans(dim1, dim2, dim3, cipher_img_trans)
#     for _ in range(5):
    cipher_img_trans = encrypt_trans1(dim1, dim2, dim3, cipher_img_trans)
    
    key = keygen(x, r, img_arr.shape[0] * img_arr.shape[1] * img_arr.shape[2])
    
    first_dim, second_dim, third_dim = decToBinary_dims(key)
    mat = get_mat(img_arr, encoding_rule1, first_dim, second_dim, third_dim)
    dim1_0, dim2_0, dim3_0 = get_dims(img_arr, mat)
    
    cipher_img = np.empty(img_arr.shape, dtype = 'uint8')
    
#     xkey, ykey, zkey = Lorenzkey(dim1_0[i], dim2_0[i], dim3_0[i], img_arr.shape[1] + 5)
#     xkey, ykey, zkey = xkey[6:], ykey[6:], zkey[6:]
    
    for i in range(img_arr.shape[0]):

        xkey, ykey, zkey = Lorenzkey(dim1_0[i], dim2_0[i], dim3_0[i], img_arr.shape[1] + 10)
        xkey, ykey, zkey = xkey[6:-5], ykey[6:-5], zkey[6:-5]
        
        j = 0
        for x in xkey:
            cipher_img[i,j,0] = cipher_img_trans[i,j,0] + int(x)
            cipher_img[i,j,0] = cipher_img[i,j,0] % 256
            j += 1

        j = 0
        for y in ykey:
            cipher_img[i,j,1] = cipher_img_trans[i,j,1] + int(y)
            cipher_img[i,j,1] = cipher_img[i,j,1] % 256
            j += 1

        j = 0
        for z in zkey:
            cipher_img[i,j,2] = cipher_img_trans[i,j,2] + int(z)
            cipher_img[i,j,2] = cipher_img[i,j,2] % 256
            j += 1
            
            
            
    meta_key = get_meta_key(cipher_img)
    print(meta_key)
    key = keygen(meta_key, r, img_arr.shape[0] * img_arr.shape[1] * img_arr.shape[2])
#     key = np.reshape(key, (img_arr.shape[0], img_arr.shape[1], img_arr.shape[2]))
    gcd_matrix = make_gcd_matrix(key)
    sum_mod = get_sum_mod(gcd_matrix)

    dim1, dim2, dim3 = trans(sum_mod)
    cipher_img_trans = cipher_img.copy()
    
#     for _ in range(3):
    cipher_img_trans = encrypt_trans(dim1, dim2, dim3, cipher_img)
#     for _ in range(5):
    cipher_img_trans = encrypt_trans1(dim1, dim2, dim3, cipher_img_trans)
                   
    return cipher_img_trans
```

```python
def decrypt_image(cipher_img, r, x):
    
    encoding_rule1 = {'00' : 'A', '11' : 'T', '01' : 'C', '10' : 'G'}
    encoding_rule2 = {'00' : 'A', '11' : 'T', '01' : 'G', '10' : 'C'}
    encoding_rule3 = {'00' : 'C', '11' : 'G', '01' : 'A', '10' : 'T'}
    encoding_rule4 = {'00' : 'G', '11' : 'C', '01' : 'A', '10' : 'T'}
    encoding_rule5 = {'10' : 'A', '01' : 'T', '00' : 'C', '11' : 'G'}
    encoding_rule6 = {'10' : 'A', '01' : 'T', '11' : 'C', '00' : 'G'}
    encoding_rule7 = {'11' : 'A', '00' : 'T', '01' : 'C', '10' : 'G'}
    encoding_rule8 = {'11' : 'A', '00' : 'T', '10' : 'C', '01' : 'G'}
    
    meta_key = get_meta_key(cipher_img)
    print(meta_key)
    key = keygen(meta_key, r, img_arr.shape[0] * img_arr.shape[1] * img_arr.shape[2])
#     key = np.reshape(key, (img_arr.shape[0], img_arr.shape[1], img_arr.shape[2]))
    gcd_matrix = make_gcd_matrix(key)
    sum_mod = get_sum_mod(gcd_matrix)

    dim1, dim2, dim3 = trans(sum_mod)
    cipher_img_trans = cipher_img.copy()
    
#     for _ in range(3):
    cipher_img_trans = decrypt_trans1(dim1, dim2, dim3, cipher_img)
#     for _ in range(5):
    cipher_img_trans = decrypt_trans(dim1, dim2, dim3, cipher_img_trans)
    
    cipher_img = cipher_img_trans.copy()
    
    key = keygen(x, r, cipher_img.shape[0] * cipher_img.shape[1] * cipher_img.shape[2])
    
    first_dim, second_dim, third_dim = decToBinary_dims(key)
    mat = get_mat(cipher_img, encoding_rule1, first_dim, second_dim, third_dim)
    dim1_0, dim2_0, dim3_0 = get_dims(cipher_img, mat)
    
    plain_img = np.empty(cipher_img.shape, dtype = 'uint8')

    for i in range(cipher_img.shape[0]):
        
        xkey, ykey, zkey = Lorenzkey(dim1_0[i], dim2_0[i], dim3_0[i], cipher_img.shape[1] + 10)
        xkey, ykey, zkey = xkey[6:-5], ykey[6:-5], zkey[6:-5]
        
        j = 0
        for x in xkey:
            plain_img[i,j,0] = cipher_img[i,j,0] - int(x)
            plain_img[i,j,0] = plain_img[i,j,0] % 256
            j += 1

        j = 0
        for y in ykey:
            plain_img[i,j,1] = cipher_img[i,j,1] - int(y)
            plain_img[i,j,1] = plain_img[i,j,1] % 256
            j += 1

        j = 0
        for z in zkey:
            plain_img[i,j,2] = cipher_img[i,j,2] - int(z)
            plain_img[i,j,2] = plain_img[i,j,2] % 256
            j += 1
            
    meta_key = get_meta_key(plain_img)
    print(meta_key)
    key = keygen(meta_key, r, cipher_img.shape[0] * cipher_img.shape[1] * cipher_img.shape[2])
    key = np.reshape(key, (img_arr.shape[0], img_arr.shape[1], img_arr.shape[2]))
    
    gcd_matrix = make_gcd_matrix(key)
    sum_mod = get_sum_mod(gcd_matrix)
    dim1, dim2, dim3 = trans(sum_mod)
    
#     for _ in range(3):
    plain_img = decrypt_trans1(dim1, dim2, dim3, plain_img)
#     for _ in range(5):
    plain_img = decrypt_trans(dim1, dim2, dim3, plain_img)
    
    return plain_img
```

```python
def get_npcr(cipher_img1, cipher_img):
    
    npcr = 0
    
    for i in range(cipher_img1.shape[0]):
        for j in range(cipher_img1.shape[1]):
            for k in range(cipher_img1.shape[2]):
                if cipher_img1[i,j,k] != cipher_img[i,j,k]:
                    npcr += 1
#     print(npcr)
    npcr = (npcr * 100)/ (cipher_img1.shape[0] * cipher_img1.shape[1] * 3)
    
    return npcr
```

```python
get_npcr(cipher_img1, cipher_img)
```

```python
import sys

def get_uaci(cipher_img1, cipher_img):
    cipher_img = cipher_img / 255
    cipher_img1 = cipher_img1 / 255
    uaci = 0.0
    for i in range(cipher_img1.shape[0]):
        for j in range(cipher_img1.shape[1]):
            for k in range(cipher_img1.shape[2]):
                uaci += abs(cipher_img1[i,j,k] - cipher_img[i,j,k])

    uaci = uaci / (cipher_img.shape[0] * cipher_img.shape[1] * 3)
    return uaci
```
