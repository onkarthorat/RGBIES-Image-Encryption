```python
def exchange(i, j):
    temp = i
    i = j
    j = temp
    
    return int(i), int(j)
```

```python
i, j = 2,3
print(i,j)
```

```python
i, j = exchange(i,j)
print(i,j)
```

```python
# cipher_img = np.empty(img_arr.shape)
def scramble(img_arr1):
    img_arr = img_arr1.copy()
    for i in range(2, img_arr.shape[0] - 2):
        for j in range(2, img_arr.shape[1] - 2):
            for k in range(3):
#                 print(i,j,k)
                img_arr[i+2, j-2, k], img_arr[i-2, j+2, k] = exchange(img_arr[i+2, j-2, k], img_arr[i-2, j+2, k])
                img_arr[i+2, j-1, k], img_arr[i+1, j+2, k] = exchange(img_arr[i+2, j-1, k], img_arr[i+1, j+2, k])
                img_arr[i+2, j, k], img_arr[i-1, j+2, k] = exchange(img_arr[i+2, j, k], img_arr[i-1, j+2, k])
                img_arr[i+2, j+1, k], img_arr[i, j+2, k] = exchange(img_arr[i+2, j+1, k], img_arr[i, j+2, k])
                img_arr[i, j+2, k], img_arr[i+2, j+2, k] = exchange(img_arr[i, j+2, k], img_arr[i+2, j+2, k])
                img_arr[i+1, j-2, k], img_arr[i-2, j+1, k] = exchange(img_arr[i+1, j-2, k], img_arr[i-2, j+1, k])
                img_arr[i+1, j-1, k], img_arr[i, j+1, k] = exchange(img_arr[i+1, j-1, k], img_arr[i, j+1, k])
                img_arr[i+1, j, k], img_arr[i-1, j+1, k] = exchange(img_arr[i+1, j, k], img_arr[i-1, j+1, k])
                img_arr[i-1, j+1, k], img_arr[i+1, j+1, k] = exchange(img_arr[i-1, j+1, k], img_arr[i+1, j+1, k])
                img_arr[i, j-2, k], img_arr[i-2, j, k] = exchange(img_arr[i, j-2, k], img_arr[i-2, j, k])
                img_arr[i, j-2, k], img_arr[i-1, j, k] = exchange(img_arr[i, j-2, k], img_arr[i-1, j, k])
                img_arr[i-2, j, k], img_arr[i, j, k] = exchange(img_arr[i-2, j, k], img_arr[i, j, k])
                img_arr[i-1, j-2, k], img_arr[i-2, j-1, k] = exchange(img_arr[i-1, j-2, k], img_arr[i-2, j-1, k])
                img_arr[i-2, j-1, k], img_arr[i-1, j-1, k] = exchange(img_arr[i-2, j-1, k], img_arr[i-1, j-1, k])
                      
                
    return img_arr
```

```python
img_arr1 = scramble(img_arr)
```
